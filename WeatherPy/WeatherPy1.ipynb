## Generate Cities List

```python
city_df = pd.read_csv("../Resource/worldcities500.csv")
cities_df = city_df[["city"]].copy()
cities_df.head()

cities_list = cities_df["city"]
print(cities_list)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather"

# Setup parameters for the API
params = {"units" : "matric",
         "appid" : weather_map_key,
         "q" : "Melbourne"
}

#Test API and display a city to show the structure of data in API
response = requests.get(url, params = params ).json()
pprint(response)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_map_key}&units={units}"
```

```python
counter = 0
Lat=[] 
Lng = [] 
Max_Temp = []
Humidity=[]
Cloudiness=[]
Wind_speed=[]
Country=[]
Date=[]
City=[]

for x in cities_list:
    print(x)
    #params["q"] = str(x)
    response = requests.get(query_url + "&q=" + str(x)).json()
                            
    try:
        print("Processing information for city number " + str(counter) + " " + x)
        Lat.append(response["coord"]["lat"])
        counter+=1
        Lng.append(response["coord"]["lon"])
        Max_Temp.append(response["main"]["temp_max"])
        Humidity.append(response["main"]["humidity"])
        Cloudiness.append(response["clouds"]["all"])
        Wind_speed.append(response["wind"]["speed"])
        Country.append(response["sys"]["country"])
        Date.append(response["dt"])
        City.append(response["name"])
    except:
        print("City not found")    
```

```python
new_df = pd.DataFrame({"City": City,
                       "Lat" : Lat,
                       "Lng" : Lng,
                       "Max Temp" : Max_Temp,
                       "Humidity" : Humidity,
                       "Cloudiness" : Cloudiness,
                       "Wind speed" : Wind_speed,
                       "Country" : Country,
                       "Date" : Date})
```

```python
#Save the dataframe into a CSV file
new_df.to_csv("../Output/cities.csv", index = False, header = True)
```
