```python
# Dependencies
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import weather_map_key, google_key
from pprint import pprint
from random import sample
```

## Generate Cities List

```python
city_df = pd.read_csv("../Resource/worldcities.csv")
cities_df = city_df[["city"]].copy()
cities_df.head()


cities_list1 = cities_df["city"].tolist()


cities_list = sample(cities_list1, 550)
cities_list
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather"

# Setup parameters for the API
params = {"units" : "matric",
         "appid" : weather_map_key,
         "q" : "Melbourne"
}

#Test API and display a city to show the structure of data in API
response = requests.get(url, params = params ).json()
pprint(response)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_map_key}&units={units}"
```

```python
counter = 0
Lat=[] 
Lng = [] 
Max_Temp = []
Humidity=[]
Cloudiness=[]
Wind_speed=[]
Country=[]
Date=[]
City=[]

for x in cities_list:
    print(x)
    #params["q"] = str(x)
    response = requests.get(query_url + "&q=" + str(x)).json()
                            
    try:
        print("Processing information for city number " + str(counter) + " " + x)
        Lat.append(response["coord"]["lat"])
        counter+=1
        Lng.append(response["coord"]["lon"])
        Max_Temp.append(response["main"]["temp_max"])
        Humidity.append(response["main"]["humidity"])
        Cloudiness.append(response["clouds"]["all"])
        Wind_speed.append(response["wind"]["speed"])
        Country.append(response["sys"]["country"])
        Date.append(response["dt"])
        City.append(response["name"])
    except:
        print("City not found")    
```

```python
new_df = pd.DataFrame({"City": City,
                       "Lat" : Lat,
                       "Lng" : Lng,
                       "Max Temp" : Max_Temp,
                       "Humidity" : Humidity,
                       "Cloudiness" : Cloudiness,
                       "Wind speed" : Wind_speed,
                       "Country" : Country,
                       "Date" : Date})
```

```python
#Save the dataframe into a CSV file
new_df.to_csv("../Output/cities.csv", index = False, header = True)
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
humidity = new_df["Humidity"]
print(humidity)
```

```python
#  Get the indices of cities that have humidity over 100%.
row_index = 0
for x in humidity:
    if x > 100:
        print("oh no, humidity is more than 100%! That's hot!")
        row_index +=1
    else:
        new_df.drop([row_index])
```

```python
Northern_Hemisphere = new_df.loc[new_df["Lat"]>=0, :]
Southern_Hemisphere = new_df.loc[new_df["Lat"]<0, :]        
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = Northern_Hemisphere["Lat"]
y_values = Northern_Hemisphere["Max Temp"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(10,0),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title("Northern Hemisphere City Latitude Vs. Max Temperature (17/03/2021)")
plt.grid()
print(f"The r-squared is: {rvalue**2}")
plt.savefig('../Output/NT_Temp_Lat.png')
plt.show()
```
